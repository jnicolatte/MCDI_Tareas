# 1B1. Práctica: Cálculo de frecuencias a nivel de un carácter
Instrucciones:


Crear un notebook para preprocesar el texto y convertirlo a minúsculas, quitar acentos y los siguientes caracteres:

`;:,.\\-\"'/()[]¿?¡!{}~<>|«»-—’\t\n\r`


```python
#librerías que utilizaremos:
import re
import unicodedata
import matplotlib.pyplot as plt
```

```python
#leemos y guardamos los textos.
file1 = open( 'text_1.txt', 'r' )
text1 = file1.read()

file2 = open( 'text_2.txt', 'r' )
text2 = file2.read()

file3 = open( 'text_3.txt', 'r' )
text3 = file3.read()

file4 = open( 'text_4.txt', 'r' )
text4 = file4.read()

file5 = open( 'text_5.txt', 'r' )
text5 = file5.read()
```

```python
#convertimos los textos a minúsculas:
text1_minus = text1.lower()

text2_minus = text2.lower()

text3_minus = text3.lower()

text4_minus = text4.lower()

text5_minus = text5.lower()
```

```python
# definimos una función para quitar los acentos:

def remover_acentos(texto):
    nfkd_form = unicodedata.normalize('NFKD',texto)
    return u"".join([c for c in nfkd_form if not unicodedata.combining(c)])
```

```python
#guardamos en nuevas variables
text1_minus_ac = remover_acentos(text1_minus)
text2_minus_ac = remover_acentos(text2_minus)
text3_minus_ac = remover_acentos(text3_minus)
text4_minus_ac = remover_acentos(text4_minus)
text5_minus_ac = remover_acentos(text5_minus)
```

```python
#quitamos caracteres especiales usando re:
text1_final = re.sub(r"[;:,.\-\"'/\(\)\[\]¿?¡!\{\}~<>|«»-—’\t\n\r]", "", text1_minus_ac)
text2_final = re.sub(r"[;:,.\-\"'/\(\)\[\]¿?¡!\{\}~<>|«»-—’\t\n\r]", "", text2_minus_ac)
text3_final = re.sub(r"[;:,.\-\"'/\(\)\[\]¿?¡!\{\}~<>|«»-—’\t\n\r]", "", text3_minus_ac)
text4_final = re.sub(r"[;:,.\-\"'/\(\)\[\]¿?¡!\{\}~<>|«»-—’\t\n\r]", "", text4_minus_ac)
text5_final = re.sub(r"[;:,.\-\"'/\(\)\[\]¿?¡!\{\}~<>|«»-—’\t\n\r]", "", text5_minus_ac)
```

## 1. Calcular las  frecuencias a nivel de un carácter, de los documentos text_1 al text_5  y generar los histogramas para cada archivo. Para generar las gráficas, ordenar los caracteres de acuerdo a su frecuencia; de mayor a menor frecuencia en el histograma. 

```python
#Definimos una función para contar los caracteres. 
def contador_caracteres(texto):
    conteo = {}
    for caracter in texto:
        conteo[caracter] = conteo.get(caracter,0) +1
    return conteo
```

```python
frecuencias_text1 = contador_caracteres(text1_final)
frecuencias_text2 = contador_caracteres(text2_final)
frecuencias_text3 = contador_caracteres(text3_final)
frecuencias_text4 = contador_caracteres(text4_final)
frecuencias_text5 = contador_caracteres(text5_final)
```

```python
#Definimos una función para generar los histogramas.
def histograma(frecuencias, texto):
    frecuencias = dict(sorted(frecuencias.items(), key=lambda item: item[1], reverse=True))
    caracteres = list(frecuencias.keys())
    valores = list(frecuencias.values())
    
    plt.bar(caracteres, valores)
    plt.xlabel("Carácter")
    plt.ylabel("Frecuencia")
    plt.title(f"Frecuencias de Caracteres en {texto}")
    plt.show()
```

```python
histograma(frecuencias_text1, "text1_final")
frecuencias_text1
```

```python
histograma(frecuencias_text2, "text2_final")
frecuencias_text2
```

```python
histograma(frecuencias_text3, "text3_final")
frecuencias_text3
```

```python
histograma(frecuencias_text4, "text4_final")
frecuencias_text4
```

```python
histograma(frecuencias_text5, "text5_final")
frecuencias_text5
```
