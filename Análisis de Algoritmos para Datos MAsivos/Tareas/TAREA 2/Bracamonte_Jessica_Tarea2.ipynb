```python
def busca_binaria(lista, identificador):
    inicio = 0
    fin = len(lista) - 1

    while inicio <= fin:
        medio = (inicio + fin) // 2
        if lista[medio] == identificador:
            return medio
        elif lista[medio] < identificador:
            inicio = medio + 1
        else:
            fin = medio - 1

    return fin+1
```

```python
def busca_secuencial(lista, identificador):
  for i, valor in enumerate(lista):
    if valor == identificador:
      return i
    elif valor > identificador:
      return i 

  return len(lista)
```
