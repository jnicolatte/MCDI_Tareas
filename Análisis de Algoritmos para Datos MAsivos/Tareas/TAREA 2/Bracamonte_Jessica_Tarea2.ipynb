```python
def busca_binaria(lista, identificador):
    inicio = 0
    fin = len(lista) - 1

    while inicio <= fin:
        medio = (inicio + fin) // 2
        if lista[medio] == identificador:
            return lista.index(identificador)
        elif lista[medio] < identificador:
            inicio = medio + 1
        else:
            fin = medio - 1

    return fin+1
```

```python
resultados = []
start = time.time()
for lista in lista_2:
    posicion_insercion = []
    for identificador in identificadores_2:
        posicion = busca_binaria(lista, identificador)
        posicion_insercion.append(posicion)
    resultados.append(posicion_insercion)
end = time.time()

tiempo = end - start


```
